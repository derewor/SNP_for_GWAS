```python
mask = (
    gff["Chromosome"].isin(snps_ri["Chromosome"]) &
    gff.apply(lambda row: any(
        (snps_ri["Chromosome"] == row["Chromosome"]) &
        (snps_ri["Start1"].between(row["Start"], row["End"]))
    ), axis=1)
)

filt_ri = gff[mask]
```

```python
gr_mlm = pd.read_csv('/content/growth_rate_mlm.csv.txt')
gr_mlm = gr_mlm[gr_mlm['p'].notna()]
gr_mlm = gr_mlm.drop(columns = ['add_effect','add_F','add_p','dom_effect',	'dom_F',	'dom_p'])
gr_mlm.head()
```

```python
gr_mlm['p'] = pd.to_numeric(gr_mlm['p'], errors='coerce')
```

```python
gr_mlm['log10(p)'] = -np.log10(gr_mlm['p']).round(2)
gr_mlm.sort_values(by='log10(p)', ascending=False)
```

```python
gr_full = pd.read_csv('/content/growth_rate_full.csv.txt')
```

```python
gr_full = gr_full.drop(columns = ['add_effect','add_F','add_p','dom_effect',	'dom_F',	'dom_p'])
gr_full = gr_full[gr_full['p'].notna()]
```

```python
# based on the p value of gr_mlm, calculate the fdr values for each marker.
gr_full['fdr'] = multipletests(gr_full['p'], method='fdr_bh')[1]
gr_full.sort_values(by='fdr', ascending=True)
```
